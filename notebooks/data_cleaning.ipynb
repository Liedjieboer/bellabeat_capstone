# Data investigation and cleaning

```python
# Import necessary libraries and packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
# Import data into pandas df's

daily_activity_df = pd.read_csv("../data/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv")
daily_calories_df = pd.read_csv("../data/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyCalories_merged.csv")
daily_intensities_df = pd.read_csv("../data/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyIntensities_merged.csv")
daily_steps_df = pd.read_csv("../data/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailySteps_merged.csv")
```

```python
def print_df_name(df):
    """A function to dynamically print the variable name of a pandas df"""
    
    name = [name for name, val in globals().items() if val is df][0]
    print(f"{name} information:/n")

def get_df_info(df):
    """A function to show the number of null values per column
        and the statistical distributions of numeric columns"""
    
    print_df_name(df)
    print(f"Null values:/n{df.isnull().sum()}")
    display(df.describe())

```

```python
# Create a list of all the df's I want to investigate

df_list = [daily_activity_df, daily_calories_df, daily_intensities_df, daily_steps_df]
```

```python
# A for loop to get the information for all the df's I want to investigate

for df in df_list:
    get_df_info(df)
```
